```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("zidensya.csv")
top_cities = df['市区町村（発生地）'].value_counts().head(10)

plt.figure(figsize=(16, 6))
bars = plt.bar(range(len(top_cities)), top_cities.values, color='mediumseagreen')

# 横軸に市区町村名を縦向きで表示、フォントサイズ小さめ
plt.xticks(ticks=range(len(top_cities)), labels=top_cities.index, rotation=90, fontsize=10)

plt.title('Top 10 Cities by Number of Incidents')
plt.xlabel('City')
plt.ylabel('Number of Incidents')
plt.grid(True, axis='both', linestyle='--', alpha=0.4)

# 棒の上に数値を表示
for i, v in enumerate(top_cities.values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
