```python
import pandas as pd

df = pd.read_csv("zidensya.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

# 市区町村（発生地）をカウントして多い順に並べる
city_counts = df["市区町村（発生地）"].value_counts().head(20)

# グラフ表示
plt.figure(figsize=(12, 6))
city_counts.plot(kind="bar")
plt.title("市区町村別の事故発生件数（上位20）")
plt.xlabel("市区町村")
plt.ylabel("件数")
plt.xticks(rotation=45, ha="right")
plt.show()
```
